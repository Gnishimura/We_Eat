```python
from pymongo import MongoClient
import pandas as pd
import numpy as np
from collections import Counter
```

```python
bizinfo_df = pd.read_pickle('bizinfo_df.pkl')
```

```python
bdf = bizinfo_df.fillna(0).replace('No', 0).replace('Yes', 1).drop(['Offers Military Discount', 'Open to All', 'Accepts Apple Pay', 'Accepts Bitcoin', 'Accepts Google Pay'], axis=1)
```

```python
bdf.columns
```

```python
def amb_cat_counts(df):
    """Create a dictionary with a list of all the categories in the 'cats' feature and their counts"""
    amb_cat_series = df['Ambience']
    amb_cat_dict = Counter()
    for row in amb_cat_series:
        if row != 0:
            temp = []
            temp = row.split(',')
            for amb_cat in temp:
                amb_cat_dict[amb_cat.strip()] += 1
    return amb_cat_dict
```

```python
ambience_cats = amb_cat_counts(bdf)
```

```python
ambience_cats
```

```python
amb_cats = list(ambience_cats)
```

```python
amb_cats
```

```python
def get_ambience_dummies(df, amb_cats):
    """Return a DataFrame of dummy columns for the ambience categories specified."""
    amb_list_col = df['Ambience'].str.split(',')
    dummied = pd.DataFrame(index=df['Ambience'].index)
    for cat in amb_cats:
        dummied['ambience_' + cat] = amb_list_col.apply(lambda x: cat in x if x != 0).astype(int)
    
    dummied_df = pd.concat([df, dummied], axis=1)

    return dummied_df
```

```python
amb_list_col = bdf['Ambience'].str.split(',')
dummied = pd.DataFrame(index=bdf['Ambience'].index)

for cat in amb_cats:
    dummied['ambience_' + cat] = amb_list_col.apply(lambda x: cat in x)
    
```
